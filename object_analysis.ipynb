```python
import numpy as np

# Extract x, y coordinates
x_values = df['x'].values
y_values = df['y'].values

# Create 2D histogram (heatmap)
heatmap, xedges, yedges = np.histogram2d(x_values, y_values, bins=(50, 50))

# Plot heatmap
plt.imshow(heatmap.T, origin='lower', cmap='hot', interpolation='nearest')
plt.title("Object Detection Heatmap")
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.colorbar(label="Detection Count")
plt.show()
```

```python
import numpy as np

df['time_seconds'] = df['Frame'] / 30  # Convert frame numbers to time (assuming 30 FPS)

# Aggregate object count over time
df_time_series = df.groupby(["time_seconds", "class_id"]).size().reset_index(name="count")

# Plot over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_time_series, x="time_seconds", y="count", hue="class_id")
plt.title("Object Frequency Over Time")
plt.xlabel("Time (Seconds)")
plt.ylabel("Number of Detections")
plt.show()
```
